```python
#Se cargan los dataframe correspondientes
df_victimas= pd.read_excel(path_homocidios,sheet_name="VICTIMAS")
df_hechos=  pd.read_excel(path_homocidios,sheet_name="HECHOS")
```

```python
#Se cambian los nombres de las columnas para ser mas intuitivos
df_hechos = df_hechos.rename(columns={'N_VICTIMAS': 'Cantidad víctimas',
                                                      'AAAA':'Año',
                                                      'MM':'Mes',
                                                      'DD':'Día',
                                                      'HORA':'Hora Completa',
                                                      'HH':'Hora',
                                                      'VICTIMA': 'Víctima',
                                                      "ACUSADO":"Acusado",
                                                      "PARTICIPANTES":"Participantes",
                                                      "TIPO_DE_CALLE":"Tipo_de_calle",
                                                      "FECHA":"Fecha",
                                                      "LUGAR_DEL_HECHO":"Lugar_del_hecho"})
```

```python
#Se elimina la columna altura debido a la gran cantidad de valores faltantes
df_hechos =df_hechos.drop('Altura', axis=1)
```

```python
#Se aplica esta funcion ya que se entiende que cuando el valor de la columna Cruce esta nulo, el siniestro no se provoco en un cruce
df_hechos["Cruce"] = df_hechos["Cruce"] .apply(lambda x: 'Cruce' if pd.notnull(x) else 'No Cruce')
```

```python
# Se elimina la columna lugar del hecho ya que contiene la misma informacion que direccion normalizada
df_hechos = df_hechos.drop(columns=['Lugar_del_hecho']).rename(columns={'Dirección Normalizada': 'Lugar_del_hecho'})
```

```python
#Se remplaza por 0 algunos valores mal cargados
df_hechos['pos x'] = df_hechos['pos x'].replace('.', 0)
df_hechos['pos y'] = df_hechos['pos y'].replace('.', 0)
df_hechos['XY (CABA)'] = df_hechos['XY (CABA)'].replace('Point (. .)', 0)
```

```python
#Remplazamos el punto por la coma

def replace_dot_with_comma(df, columns):

    for column in columns:
        df[column] = df[column].astype(str).str.replace('.', ',')
    return df

# Aplicar la función al DataFrame
columns_to_replace = ['pos x', 'pos y']
df_hechos = replace_dot_with_comma(df_hechos, columns_to_replace)
```

```python
#Se modifican los nombres para que sean mas intuitivos
df_victimas = df_victimas.rename(columns={"ID_hecho":"ID_victima",  
                                          'AAAA':'Año',
                                            'MM':'Mes',
                                            'DD':'Día',
                                            'VICTIMA':'Víctima',
                                            "SEXO":"Sexo",
                                            "FECHA_FALLECIMIENTO":"Fecha_de_fallecimiento",
                                            "ROL":"Rol",
                                            "FECHA":"Fecha",
                                            "EDAD":"Edad"})
```

```python
#Remplaza los valores SD por NaN Y despues los remplaza por el valor mas frecuente
def ingresa_valor_frecuente(df, columna):
    
    df[columna] = df[columna].replace('SD', pd.NA)
    valor_frecuente = df[columna].mode().iloc[0]
    
    df[columna] = df[columna].fillna(valor_frecuente)
    
    return df

df_victimas = ingresa_valor_frecuente(df_victimas, 'Sexo')
df_victimas= ingresa_valor_frecuente(df_victimas, 'Rol')
df_victimas= ingresa_valor_frecuente(df_victimas, 'Víctima')
```

```python
#Remplaza todos los SD por NaN y luego por el valor promedio
df_victimas['Edad'] = df_victimas['Edad'].replace('SD', np.nan)
Promedio = df_victimas["Edad"].mean()

df_victimas['Edad'] = df_victimas['Edad'].fillna(Promedio)
```

```python
#Remplaza los valores SD de Fecha_de_fallecimiento por los valores de Fecha,suponiendo que murieron ese mismo dia
def replace_sd_with_date(df, source_col, target_col, sd_value="SD"):

    df[target_col] = df.apply(lambda row: row[source_col] if row[target_col] == sd_value else row[target_col], axis=1)
    return df

df_victimas = replace_sd_with_date(df_victimas, 'Fecha', 'Fecha_de_fallecimiento')
```

```python
#Convertimos la columna fecha al tipo datatime
df_victimas['Fecha'] = pd.to_datetime(df_victimas['Fecha'])
```

```python
#Convertimos la columna Fecha_de_fallecimiento al tipo datatime
df_victimas['Fecha_de_fallecimiento'] = pd.to_datetime(df_victimas['Fecha_de_fallecimiento'])
```
